```python
# --- Data Cleaning ---
# Removing unwanted columns
columns_to_remove = ['mid', 'venue', 'batsman', 'bowler', 'striker', 'non-striker']
df.drop(labels=columns_to_remove, axis=1, inplace=True)
```

```python
actual_teams=['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals','Mumbai Indians',  'Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
df=df[(df['bat_team'].isin(actual_teams)) & (df['bowl_team'].isin(actual_teams))]
```

```python
df['bat_team'].unique()
```

```python
df_after_5=df[df['overs']>=5]
```

```python
df_after_5['date']=pd.to_datetime(df_after_5['date'])
```

```python
#data processing
dummy=pd.get_dummies(df_after_5[['bat_team','bowl_team']],drop_first=True)
df_after_5=df_after_5.join(dummy)
```

```python
df_after_5.columns
```

```python
df_after_5=df_after_5.loc[:,['date','bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
       'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians',
       'bowl_team_Rajasthan Royals', 'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad', 'runs', 'wickets', 'overs',
       'runs_last_5', 'wickets_last_5','total' ] ]
```

```python
x_train=df_after_5[df_after_5['date'].dt.year<=2016]
x_test=df_after_5[df_after_5['date'].dt.year>2016]

y_train=x_train['total']
y_test=x_test['total']
```

```python
x_train.drop('date',inplace=True,axis=1)
x_test.drop('date',inplace=True,axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
lr=LinearRegression()

def fitting(ob):
    ob.fit(x_train,y_train)
    pred_lr=ob.predict(x_test)
    print(r2_score(y_test,pred_lr))
    print('MAE',mean_absolute_error(y_test,pred_lr))
    print('MsE',mean_squared_error(y_test,pred_lr))
    sns.distplot(y_test-pred_lr)
fitting(lr)    
```

```python
dtc=DecisionTreeRegressor()
fitting(dtc)
from sklearn.model_selection import RandomizedSearchCV

max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 40, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
rand_grid={'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

rv_dtc=RandomizedSearchCV(dtc,param_distributions=rand_grid,scoring='neg_mean_squared_error',n_iter=6,cv=3,random_state=42, n_jobs = 1)
```

```python
fitting(rv_dtc)
```

```python
#Ridge Regression

rdg=Ridge()
parametres={'alpha':[1e-10,1e-5,1e-3,1e-1,1,3,10]}
rdg_reg=GridSearchCV(rdg,parametres,scoring='neg_mean_squared_error',cv=5,)
rdg_reg.fit(x_train,y_train)
```
